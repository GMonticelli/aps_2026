```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sig
```

## Generación de señal de prueba

```python
def mi_senoidal(vmax=1, dc=0, f0=1, n=1000, fs=1000, ph=0):

    ts = 1/fs
    tt = np.linspace(0, (n-1)*ts, n)
    xx = dc + vmax*np.sin(2*np.pi*f0*tt + ph)

    return tt, xx

# Parámetros
n = 1000
fs = 1000

tt, xx = mi_senoidal(n=n, fs=fs)

plt.plot(tt, xx)
plt.title("Señal senoidal")
plt.grid()
plt.show()
```

## Convolución con delta

Se analiza la convolución de una señal con una delta desplazada.  
Teóricamente, se espera obtener la misma señal desplazada.

```python
# Genero delta
n0 = 300
dd = np.zeros(n)
dd[n0] = 1

yy = (1/n)*sig.convolve(xx, dd)

plt.plot(yy)
plt.title("Convolución con delta")
plt.grid()
plt.show()
```

Se observa que la señal resultante es la misma señal original desplazada.

## Correlación

Se calcula la correlación de una señal utilizando convolución con una versión invertida (flip).

```python
# Ruido
ruido = np.random.normal(0,1,n)

yy_corr = (1/n)*sig.convolve(ruido, np.flip(ruido))

plt.plot(yy_corr)
plt.title("Autocorrelación del ruido")
plt.grid()
plt.show()
```

Se observa que la correlación presenta un máximo en el origen.

## Cuantización

Se cuantiza la señal utilizando una cantidad finita de bits.

```python
B = 3       # bits
Vfs = 3     # volts
q = Vfs / (2**B)

xxq = np.round(xx / q)

plt.plot(xxq)
plt.title("Señal cuantizada")
plt.grid()
plt.show()
```

La cuantización introduce error debido a la discretización en niveles.
